```python
import dotenv
dotenv.load_dotenv()
from openai import AsyncOpenAI

transcript = """
911 Dispatcher: 911, what’s your emergency?

Caller: Hi, I... I’m bleeding really badly. I need help.

911 Dispatcher: Okay, stay calm. Can you tell me your name and location?

Caller: My name is John. I’m at 123 Elm Street, Apartment 4B.

911 Dispatcher: John, are you alone or is there anyone with you?

Caller: I’m alone. It’s just me.

911 Dispatcher: Can you tell me where you’re bleeding from?

Caller: It’s my arm. I cut it on a piece of glass, and it’s bleeding a lot.

911 Dispatcher: How deep is the cut, John? Is it still bleeding heavily?

Caller: Yes, it’s pretty deep and it won’t stop bleeding.

911 Dispatcher: Okay, John, I need you to apply pressure to the wound. Do you have a clean cloth or towel nearby?

Caller: Yes, I have a towel.

911 Dispatcher: Great. Press the towel firmly against the wound and keep it there. Raise your arm above your heart level if you can. This will help slow the bleeding.

Caller: Okay, I’m doing it.

911 Dispatcher: Good. Keep applying pressure. I’m dispatching an ambulance to your location now. Stay on the line with me until they arrive, okay?

Caller: Alright, thank you.

911 Dispatcher: You’re doing great, John. Is the bleeding slowing down at all?

Caller: It’s still bleeding, but not as much.

911 Dispatcher: That’s good to hear. Keep the pressure on. Help is on the way. Do you feel lightheaded or dizzy?

Caller: A little bit, yeah.

911 Dispatcher: Try to stay as calm as possible and keep breathing steadily. Is there anything else you can tell me about the situation?

Caller: No, that’s it. Just the cut.

911 Dispatcher: Okay, John. The ambulance should be there very soon. Keep applying pressure and stay calm. You’re going to be alright.

Caller: Thank you.

911 Dispatcher: You’re welcome. Just hang in there. Help is almost there.

"""

```

```python
async def quick_summary(message: str):
    client = AsyncOpenAI()
    system_message = """
    You are an ai system to evaluate 911 transcripts. You should analyze the calls for their severity, type, title, summary, and location
    Output a json in the following format:
    {
        "severity": "low" | "medium" | "high" | "critical",
        "type": "fire" | "hospital" | "police",
        "name": str,
        "title": str
        "summary": str,
        "location": str
    }
    """

    summary = await client.chat.completions.create(
        model="gpt-4o",
        response_format={"type": "json_object"},
        messages=[
            {"role": "system", "content": system_message},
            {"role": "user", "content": message},
        ],
    )
    return summary.choices[0].message.content

print(await quick_summary(transcript))
```
